```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
modelnumber = 10
```

```python
mmd_lbols = np.loadtxt(f'evol_indicators_lbol.tbl', skiprows=0).T
mmd_tbols = np.loadtxt(f'evol_indicators_tbol.tbl', skiprows=0).T
cjb_lbols = np.loadtxt(f'lbols_model{modelnumber:02n}.tab', skiprows=1).T
cjb_tbols = np.loadtxt(f'tbols_model{modelnumber:02n}.tab', skiprows=1).T
```

```python
def pdif(a,b):
    return np.abs(a-b)/((a+b)/2)

def dif(a,b):
    return a/b-1
```

```python
incls = [5,15,25,35,45,55,65,75,85]
```

```python
fig, axs = plt.subplots(nrows=3,ncols=3,figsize=(20,20))

for i in range(len(incls)):

    ax = axs.ravel()[i]
    
    
    ax.scatter(mmd_lbols[0], mmd_lbols[i+2], c='tab:blue', s=8, label='radmc 2d')
    ax.scatter(cjb_lbols[0], cjb_lbols[i+2], c='orange', s=8, label='radmc 3d')
    #ax.scatter(1e6*mmd_tbols[0][1:], dif(mmd_tbols[i+2][1:], cjb_tbols[i+1]), c='tab:blue', s=16)

    ax.set_ylabel('L_bol (L$_\odot$)')
    ax.set_xlabel('Time (Myr)')
    ax.set_yscale('log')
    ax.set_title(f'inclination {incls[i]}')

    ax.legend()

# plt.savefig('figs/Lbols.eps', bbox_inches='tight')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3,ncols=3,figsize=(20,20))

for i in range(len(incls)):

    ax = axs.ravel()[i]
    
    
    ax.scatter(mmd_tbols[0], mmd_tbols[i+2], c='tab:blue', s=8, label='radmc 2d')
    ax.scatter(cjb_tbols[0], cjb_tbols[i+2], c='orange', s=8, label='radmc 3d')
    #ax.scatter(1e6*mmd_tbols[0][1:], dif(mmd_tbols[i+2][1:], cjb_tbols[i+1]), c='tab:blue', s=16)

    ax.set_ylabel('T_bol (K)')
    ax.set_xlabel('Time (Myr)')
    ax.set_yscale('log')
    ax.set_title(f'inclination {incls[i]}')

    ax.legend()

ax.axvline(cjb_tbols[0][430], c='r')
# plt.savefig('figs/Tbols.eps', bbox_inches='tight')
plt.show()
```
